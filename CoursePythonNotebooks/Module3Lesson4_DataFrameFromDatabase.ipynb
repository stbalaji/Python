## Example 3 : Joins and combining multiple datasets

```python
df_playlists = pd.read_sql_query("SELECT * FROM playlists", conn)
df_playlists
```

```python
df_playlist_track = pd.read_sql_query("SELECT * FROM playlist_track", conn)
df_playlist_track
```

```python
df_playlist_track = pd.read_sql_query("SELECT * FROM playlist_track", conn)
df_playlist_track
```

```python
df_tracks = pd.read_sql_query("SELECT * FROM tracks", conn)
df_tracks.describe()
```

### Select a subset of columns for illustration of join

```python
# Dataframe for One Playlist slice  - example 17 which is 'Heavy Metal Classic'
cond = df_playlist_track['PlaylistId'] == 17
df_one_playlist = df_playlist_track[cond]
df_one_playlist
```

```python
# Pull Subset of columns from the Track like just Composer name, song time
df_tracks_composer = df_tracks[['Composer','TrackId', 'Name','Milliseconds']]
df_tracks_composer 
```

### Join Data - Find all Playlists in which ACDC appears and we want to know the type of Playlist

```python
df_one_playlist
```

```python
df_tracks_composer.info()
```

```python
cond = df_tracks_composer['TrackId'] = 1335
df_tracks_composer[cond]
```

```python
df_playlist_composer = pd.merge(df_one_playlist, df_tracks_composer, on="TrackId")
df_playlist_composer
```

```python
cond = df_tracks['Composer'] == 'Metallica'
df_tracks_Metallica = df_tracks[cond]
df_tracks_Metallica = df_tracks_Metallica[['Composer','TrackId', 'Name','Milliseconds']]
df_tracks_Metallica
```

### Join Data -  Outer Join e.g. What songs of metallic are not in the "Heavy Metal Classic" playlist

```python
# Equi Join 
df_composer_playlist = pd.merge(df_tracks_Metallica, df_one_playlist, on="TrackId")
df_composer_playlist 
```

```python
# Note the two tables have TrackId as a common column. This will be used to combine the tables - join
df_details = pd.merge(df_tracks, df_playlist_track)
df_details.describe() 
```

```python
# Left Outer Join Join 
df_composer_playlist = pd.merge(df_tracks_Metallica, df_one_playlist, on= 'TrackId',how='left')
df_composer_playlist
```

```python
# We could also specificy columns to be used to merging
df_details_trackId = pd.merge(df_tracks, df_playlist_track, on= 'TrackId',how='left')
df_details_trackId.describe() 
```

### Note the Name of the Playlist as a column and name of Track as a column but data has been merged on Playlistid

```python
df_details_new = pd.merge(df_playlists, df_details_trackId, on= 'PlaylistId')
df_details_new 
```

```python
# Look at all playlist with moviers Name_x
cond = df_details_new['Composer'] == 'AC/DC'
df_details_new[cond]
```
