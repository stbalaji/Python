## Step 3 : waves and Attributes

```python
# Look at Log of values
num_of_points = 100
data_x = np.linspace(1, 25, num_of_points)       # Start, End and number of points
data_ysin = np.sin(data_x)                       # Sin function
data_ycos = np.cos(data_x)

plt.plot(data_x,data_ysin) 
plt.plot(data_x,data_ycos) 

plt.show
```

### Specifying line attributes 
### Color : one of the parameters to plot function - 'b' (blue), 'g' (green), 'r' (red), 'm' (magenta), 'k' (black), 'c' (cyan), ' y' (yellow), 'w' (white)

### linewidth : specify a number for the width

### linestyle  : 'solid', 'dashed', 'dotted', 'dashdot'

```python
plt.plot(data_x,data_ysin, 'r', linewidth = 6, linestyle = 'dashed') 
plt.plot(data_x,data_ycos, 'b', linewidth = 2, linestyle = 'solid') 

plt.show
```

### Datapoints are referred to as markers
### You can specify type of market

```python
plt.plot(data_x,data_ysin, 'r', linewidth = 2, linestyle = 'dashed', marker = '*', markersize = 6) 
plt.plot(data_x,data_ycos, 'b', linewidth = 2, linestyle = 'solid', marker = 'o') 
plt.show
```

### Use a Pandas Series object 

```python
series = pd.Series({'low':10, 'medium': 20, 'high': 70, 'very high':100, 'Super High':200}) 
plt.bar( series.index, series.values)  
# plt.bar( series.index, series.values, width = 0.5)  # controlling the widths of bars
# plt.bar( series.index, series.values, width = 0.5, color = 'red')  # controlling the color as well
# plt.bar( series.index, series.values, width = 0.5, color = ['r', 'b', 'g','y','k'])  # specifying color for individual bars
```

```python
plt.bar(series.index, series.values, width = 0.5, color = 'red')  # controlling the color as well
plt.bar(series.index, series.values*1.5, width = 0.5, color = 'blue')  # controlling the color as well
plt.show()
```

```python
plt.barh(series.index, series.values, color = 'red')  # controlling the color as well
plt.show()
```

### Creating a Pie Chart

```python
s_countries  = ['USA', 'UK', 'Germany', 'India','Australia']
s_population = [6, 9, 3.5, 19, 0.5]
plt.pie(s_population, labels=s_countries)
plt.show
```

```python
s_countries  = ['USA', 'UK', 'Germany', 'India','Australia']
s_population = [6, 9, 3.5, 19, 0.5]
explodes = (0, 0.1, 0, 0.1,0)           # Explode 2nd and 4th values
plt.pie(s_population, labels=s_countries, explode=explodes, autopct='%1.1f%%', shadow=True)
plt.show
```
