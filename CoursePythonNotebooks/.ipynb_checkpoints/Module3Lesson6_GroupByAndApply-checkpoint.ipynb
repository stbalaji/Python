####  > GroupBy and Aggregation Examples

```python
# Count by value
df_customers = pd.read_sql_query("SELECT * FROM customers", conn)
df_customers.groupby('State').State.count()
```

```python
df_grouped_state = df_customers.groupby('State')
```

```python
df_grouped_state.State.count() > 1         # Condition
```

####  > Let us look at Numerical columns and calculation examples

```python
df_invoices = pd.read_sql_query("SELECT * FROM invoices", conn)
df_invoices
```

```python
df_invoices.groupby('CustomerId').Total.count()  # Number of invoices
```

```python
#### Print Invoice totals by Customer
```

```python
df_invoices.groupby('CustomerId').Total.sum()    # Example for customerid 4 the value to be 39.32 (EUR?), for 57 value is 46.62
```

#### Print or get list of maximum invoices per customer

```python
df_invoices.groupby('CustomerId').Total.max()    # Example for customerid 4 the value to be 39.32 (EUR?), for 57 value is 46.62
```

#### Datatime components and adding columns as part of data preparation - Example add Year and Month

```python
df_invoices.InvoiceDate = pd.to_datetime(df_invoices['InvoiceDate'])
df_invoices['Year'] = df_invoices['InvoiceDate'].dt.year
df_invoices['Month'] = df_invoices['InvoiceDate'].dt.month
df_invoices
```

#### Think of each groupby as a slice of dataframe containing only matching subset of values.
#### This subset can be accessed using .apply. Example we want to print the first invoice per customer.

```python
df_invoices_results = df_invoices[['CustomerId','InvoiceId', 'Year', 'Month', 'Total']]            # Usually it is good to only filter columns you may be interested in 
df_invoices_results
```

### Structure of the Grouped Objected will be like (key, dataframe) where key is the grouped value

![GroupBy](dataframe_grouped.png)

```python
df_grouped_invoice_customerid = df_invoices_results.groupby('CustomerId') 
df_grouped_invoice_customerid
```

```python
for customer_id, customer_id_df in df_grouped_invoice_customerid:
    print(customer_id)
    print(customer_id_df)
```

```python
df_grouped_invoice_customerid.apply(lambda df_invoices: df_invoices.InvoiceId.iloc[0]) # for Customer 6 it should be 46
```

### Query  : What is the invoice total per customer

```python
df_grouped_invoice_customerid['Total'].sum()
```

```python
df_grouped_invoice_customerid['Total'].agg([np.mean, min, max])  # Take one example and you can verify like customerid 6
```

```python
# Take one example like customer id 6 for reference or validation
cond = df_invoices.CustomerId  == 6
df_invoices[cond]
```

```python
### We can apply different functions to different columns - using dictionary as an argument to agg method.
### Query 2 : Let us say we want find out statistics on today but also find out the frequent month invoices are raised for a customer
```

```python
df_grouped_invoice_customerid[['Total', 'Month']].agg({'Total':[np.mean, min, max], 'Month':[np.median]})  # Take one example and you can verify like customerid 6
```

### Group by Two columns : Example by Customer and by Year

```python
df_grouped_invoice_2cols  = df_invoices_results.groupby(['CustomerId', 'Year']) 
for i in df_grouped_invoice_2cols:
    print(i)
```
